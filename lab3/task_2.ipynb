```python
import random
import numpy as np
from Kupyna.KupynaRealization import KupynaRealization
random.seed(48792357293)
```

```python
def modify_single_bit(b):
    index = random.randint(0, len(b) - 1)
    bit_pos = random.randint(0, 7)
    mask = 1 << bit_pos
    b[index] = b[index] ^ mask
    return b

def hash_difference(h1, h2):
    return sum((x ^ y).bit_count() for x, y in zip(h1, h2))
k = KupynaRealization(256)

N = 256
trials = 10
sample = []

alphabet = list(map(chr, range(ord("А"), ord("я")))) + ["і", "ї", "є"]
def randomLetter():
    letter = alphabet[random.randint(0, len(alphabet) - 1)]
    return letter

for x in range(trials):
    pangrama = "".join([randomLetter() for _ in range(N)])
    print(pangrama)
    b = bytearray(pangrama.encode("utf8"))
    hash_b = k.messageHashing(b)
    b = modify_single_bit(b)
    hash_b1 = k.messageHashing(b)
    diff = hash_difference(hash_b, hash_b1)
    print(diff)
    sample.append(diff)
sample = np.array(sample)
print(sample)
print()
print(f"Довжина Хешу: {k.N}")
print(f"Довжина повідомлення в бітах: {(N * 8)}")
print(f"(Середнє(std)) Бітові зміни для модифікації одного вхідного біта: {sample.mean()}")
print(f"(Розбіжність(std)) Бітові зміни для модифікації одного вхідного біта: {sample.std()}")

# str_ = "000102030405060708090A080C0D0E0F101112131415161718191A181C1D1E1F202122232425262728292A2B2C2D2E2F303132333435363738393A383C3D3E3F404142434445464748494A4B4C4D4E4F505152535455565758595A5B5C5D5E5F606162636465666768696A6B6C6D6E6F707172737475767778797A7B7C7D7E7F"
# bytee = bytearray("".join([str_[_] for _ in range(len(str_))]).encode("utf-8"))
# a = k.messageHashing(bytee)
# print(a)
# print(a.hex())
```
